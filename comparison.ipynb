```python
import json
```

```python
with open('val_vim.json') as f:
    json_vim = json.load(f)
with open('val_vit.json') as f:
    json_vit = json.load(f)
```

```python
import matplotlib.pyplot as plt

# comparison of ViT and ViM losses over epochs
plt.figure(figsize=(8, 6))
plt.plot(json_vit['val_losses'], label='ViT')
plt.plot(json_vim['val_losses'], label='ViM')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison')
plt.legend()
plt.grid(True)
plt.savefig('loss_comparison.png')
plt.show()

# comparison of ViT and ViM F1 scores over epochs
plt.figure(figsize=(8, 6))
plt.plot(json_vit['val_f1s'], label='ViT')
plt.plot(json_vim['val_f1s'], label='ViM')
plt.xlabel('Epochs')
plt.ylabel('Validation F1 Score')
plt.title('Validation F1 Score Comparison')
plt.legend()
plt.grid(True)
plt.savefig('f1_comparison.png')
plt.show()

# bar graph comparing training/validation times
models = ['ViT', 'ViM']
times = [json_vit['time'], json_vim['time']]

plt.figure(figsize=(8, 6))
plt.bar(models, times)
plt.xlabel('Model')
plt.ylabel('Time (seconds)')
plt.title('Training+Validation Time Comparison')
plt.savefig('time_comparison.png')
plt.show()
```
